```python
import pandas as pd
import numpy as np
df = pd.read_excel('/content/Result-Experiment-2 - Copy.xlsx')
df
```

```python
grade_mapping = {
    'O': 10, 'A+': 9, 'A': 8, 'B+': 7, 'B': 6,
    'C': 5, 'D': 4, 'F': 3,'AB':0,'P':4
}
df_encoded = df.copy()
```

```python
import pandas as pd
import warnings

# Ignore future warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

# Step 2: Make a copy to avoid modifying the original
df_encoded = df.copy()

# Step 3: Grade mapping
grade_mapping = {
    'O': 10,
    'A+': 9,
    'A': 8,
    'B+': 7,
    'B': 6,
    'C+': 5
}

# Step 4: Detect subject columns (exclude unnamed or ID/name columns)
subject_columns = [col for col in df.columns if not col.startswith('Unnamed') and col != 'Name']

# Step 5: Encode grades
for col in subject_columns:
    df_encoded[col] = df_encoded[col].map(grade_mapping).fillna(df_encoded[col])

# Step 6: Convert to numeric for mean calculation
for col in subject_columns:
    df_encoded[col] = pd.to_numeric(df_encoded[col], errors='coerce')

# Step 7: Calculate mean
df_encoded['Mean_Grade'] = df_encoded[subject_columns].mean(axis=1, skipna=True)

# Step 8: Classify based on mean
def classify(mean):
    if 9 <= mean <= 10:
        return 1
    elif 6 <= mean < 9:
        return 2
    elif 4 <= mean < 6:
        return 3
    else:
        return 4

df_encoded['Class'] = df_encoded['Mean_Grade'].apply(classify)

# Step 9: Save result locally (within Colab environment)
output_file = "Encoded_Grades_Final.xlsx"
df_encoded.to_excel(output_file, index=False)

# Step 10: Show result in notebook
df_encoded.head()  # Displays first few rows
```

```python
# Assuming df_encoded is already defined and contains encoded grades

# Convert encoded grade columns to numeric
for col in subject_columns:
    df_encoded[col] = pd.to_numeric(df_encoded[col], errors='coerce')

# Calculate mean grade across subjects
df_encoded['Mean_Grade'] = df_encoded[subject_columns].mean(axis=1, skipna=True)

# Classify based on mean grade
def classify(mean):
    if 9 <= mean <= 10:
        return 1
    elif 6 <= mean < 9:
        return 2
    elif 4 <= mean < 6:
        return 3
    else:
        return 4

df_encoded['Class'] = df_encoded['Mean_Grade'].apply(classify)

# Save final file
final_file_path = "/content/Encoded_Grades.xlsx"
df_encoded.to_excel(final_file_path, index=False)

print(f"Final file with mean and class saved at: {final_file_path}")

# Display updated DataFrame
df_encoded.head()
```
