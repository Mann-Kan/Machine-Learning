```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("titanic.csv")

# Display the first few rows
df.head()
```

```python
# 1. Find Shape and Number of Columns
print("Shape of Dataset:", df.shape)
print("Number of Columns:", df.shape[1])
```

```python
# 2. Display Dataset Info
print("\nDataset Info:")
print(df.info())
```

```python
# 3. Describe Dataset (Statistical Summary)
print("\nStatistical Summary:")
print(df.describe())
```

```python
# 4. Find Null Values and Sum of Null Values
print("\nNull Values Per Column:")
print(df.isnull().sum())

print("\nTotal Missing Values:", df.isnull().sum().sum())
```

```python
# Display rows where Age is null
null_age_rows = df[df['Age'].isnull()]
print(null_age_rows)

# Print row indices where Age is null
null_age_rows = df[df['Age'].isnull()].index
print("Rows with null Age values:", null_age_rows.tolist())

# Display rows where 'Age' is null
null_age_rows = df[df['Age'].isnull()]

# Print the result
print("Rows where Age is null:")
print(null_age_rows)

print(null_age_rows[['PassengerId', 'Name', 'Sex', 'Survived', 'Age']])
```

```python
# 5. Find Unique Values
print("\nUnique Values Per Column:")
print(df.nunique())
```

```python
# 6. Plot Histograms for Numeric Columns
df.hist(figsize=(12, 8), bins=30, edgecolor='black')
plt.suptitle("Histograms of Numeric Columns")
plt.show()
```

```python
# 7. Scatterplot (Age vs Fare)
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Age"], y=df["Fare"], hue=df["Survived"])
plt.title("Scatterplot: Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
# 8. Find Pairwise Correlation
print("\nPairwise Correlation Matrix:")
print(df.corr(numeric_only=True))
```

```python
# 9. Heatmap (Correlation Matrix)
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Boxplots for outlier detection
# Select numeric columns
df_numeric = df.select_dtypes(include=[np.number])

# Calculate number of rows and columns for subplots
num_cols = len(df_numeric.columns)
num_rows = (num_cols + 2) // 3  # Calculate rows, ensuring at least 1

plt.figure(figsize=(8, 4))  # Adjust figure height based on rows

for i, col in enumerate(df_numeric.columns, 1):
    plt.subplot(num_rows, 3, i)  # Use calculated rows and 3 columns
    sns.boxplot(x=df_numeric[col], color="lightblue")
    plt.title(f"Boxplot of {col}")

plt.tight_layout()
plt.show()
```

```python
# Handle Outliers (Optional: Capping extreme values at 99th percentile)
df["Fare"] = np.where(df["Fare"] > df["Fare"].quantile(0.99), df["Fare"].quantile(0.99), df["Fare"])
df["Age"] = np.where(df["Age"] > df["Age"].quantile(0.99), df["Age"].quantile(0.99), df["Age"])
```

```python
# Find the number of missing values
missing_values = df.isnull().sum().sum()
print("Total Missing Values:", missing_values)
```

```python
# Find number of people survived
total_survived = df['Survived'].sum()
print("Total People Survived:", total_survived)
```

```python
# Find how many males and females survived
gender_survival = df.groupby('Sex')['Survived'].sum()
print("\nNumber of Males and Females who Survived:")
print(gender_survival)
```

```python
# Find the average age of people who survived
average_age_survived = df[df['Survived'] == 1]['Age'].mean()
print("\nAverage Age of Survived People:", round(average_age_survived, 2))
```

```python
# Find the minimum and maximum age of people who survived
min_age_survived = df[df['Survived'] == 1]['Age'].min()
max_age_survived = df[df['Survived'] == 1]['Age'].max()
print("\nMinimum Age of Survived People:", min_age_survived)
print("Maximum Age of Survived People:", max_age_survived)
```

```python
# Find the survival count with respect to class
class_survival = df.groupby('Pclass')['Survived'].sum()
print("\nSurvival Count per Class:")
print(class_survival)
```

```python
# How many passengers have siblings/spouses on the Titanic
passengers_with_siblings_spouses = df[df['SibSp'] > 0].shape[0]
print("\nNumber of Passengers with Siblings/Spouses on Titanic:", passengers_with_siblings_spouses)
```

```python
# How many passengers have their parents/children on the Titanic
passengers_with_parents_children = df[df['Parch'] > 0].shape[0]
print("Number of Passengers with Parents/Children on Titanic:", passengers_with_parents_children)
```

```python
import pandas as pd

# Checking initial missing values in 'Age' column
print("Missing Age values breakdown before filling:")
print(df.groupby(['Sex', 'Survived'])['Age'].apply(lambda x: x.isnull().sum()))

# Calculate mean Age separately for Survived & Non-Survived Males and Females
mean_ages = df.groupby(['Sex', 'Survived'])['Age'].mean()

# Display the calculated averages
print("\nCalculated Average Age for each group:")
print(mean_ages)

# Fill missing Age values based on Gender & Survival Status
df.loc[(df['Sex'] == 'male') & (df['Survived'] == 1) & (df['Age'].isnull()), 'Age'] = mean_ages['male', 1]
df.loc[(df['Sex'] == 'male') & (df['Survived'] == 0) & (df['Age'].isnull()), 'Age'] = mean_ages['male', 0]
df.loc[(df['Sex'] == 'female') & (df['Survived'] == 1) & (df['Age'].isnull()), 'Age'] = mean_ages['female', 1]
df.loc[(df['Sex'] == 'female') & (df['Survived'] == 0) & (df['Age'].isnull()), 'Age'] = mean_ages['female', 0]

# Checking missing values after filling
print("\nMissing Age values breakdown after filling:")
print(df.groupby(['Sex', 'Survived'])['Age'].apply(lambda x: x.isnull().sum()))
```
